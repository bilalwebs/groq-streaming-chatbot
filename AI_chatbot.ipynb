```python
import gradio as gr
from groq import Groq
from google.colab import userdata

# =========================
# SETUP CLIENT (SECURE)
# =========================
api_key = userdata.get("GROQ_API_KEY").strip()

client = Groq(api_key=api_key)

# =========================
# CHAT FUNCTION
# =========================
def chat(message, history):
    try:
        messages = [{"role": "system", "content": "You are a helpful AI assistant."}]

        # Add history
        for user, bot in history:
            messages.append({"role": "user", "content": user})
            messages.append({"role": "assistant", "content": bot})

        # Add current message
        messages.append({"role": "user", "content": message})

        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=messages,
            temperature=0.7
        )

        return response.choices[0].message.content

    except Exception as e:
        return f"⚠️ Error: {str(e)}"

# =========================
# GRADIO UI
# =========================
demo = gr.ChatInterface(
    fn=chat,
    title="🤖 Groq AI Chatbot",
    description="Ask anything...",
)

# =========================
# LAUNCH
# =========================
demo.launch()
```

```python
import gradio as gr
from groq import Groq
from google.colab import userdata

# =========================
# SECURE CLIENT
# =========================
api_key = userdata.get("GROQ_API_KEY").strip()
client = Groq(api_key=api_key)

# =========================
# STREAMING FUNCTION
# =========================
def chat(message, history):
    messages = [{"role": "system", "content": "You are a helpful AI assistant."}]

    # Add history
    for user, bot in history:
        messages.append({"role": "user", "content": user})
        messages.append({"role": "assistant", "content": bot})

    messages.append({"role": "user", "content": message})

    # 🔥 STREAM ENABLE
    stream = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=messages,
        stream=True
    )

    partial_response = ""

    # 🔄 STREAM LOOP
    for chunk in stream:
        if chunk.choices[0].delta.content:
            partial_response += chunk.choices[0].delta.content
            yield partial_response  # 👈 MAGIC LINE

# =========================
# UI
# =========================
demo = gr.ChatInterface(
    fn=chat,
    title="⚡ Streaming Groq Chatbot",
    description="Real-time AI typing...",
)

demo.launch()
```

```python
import gradio as gr
from groq import Groq
from google.colab import userdata
import time

# =========================
# CLIENT
# =========================
api_key = userdata.get("GROQ_API_KEY").strip()
client = Groq(api_key=api_key)

# =========================
# SLOW STREAM FUNCTION
# =========================
def chat(message, history):
    messages = [{"role": "system", "content": "You are a helpful AI assistant."}]

    for user, bot in history:
        messages.append({"role": "user", "content": user})
        messages.append({"role": "assistant", "content": bot})

    messages.append({"role": "user", "content": message})

    stream = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=messages,
        stream=True
    )

    partial = ""

    for chunk in stream:
        if chunk.choices[0].delta.content:
            token = chunk.choices[0].delta.content
            partial += token

            time.sleep(0.05)  # 👈 SPEED CONTROL (change this)

            yield partial + "▌"  # cursor effect

    yield partial  # final clean output

# =========================
# UI
# =========================
demo = gr.ChatInterface(
    fn=chat,
    title="🐢 Slow Streaming Chatbot",
    description="Human-like typing effect",
)

demo.launch()
```
